## 6)Indian PAN card format follows the following formats -

AYEPC7894X    
ABCDE9999YTake user input for PAN_CARD and validate as per the above example.

```python
pan=input()
if pan[0:5].isalpha() and pan[5:9].isdigit() and pan[-1].isalpha and len(pan) == 10 and pan[5:9] != "0000" :  
  print(f" Your pan number :  {pan}  is valid")
else:
   print(f" Your pan number :  {pan}  is not valid")
```
